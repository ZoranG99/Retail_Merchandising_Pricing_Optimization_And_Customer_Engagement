# INSPECTION

```python
import os
import pandas as pd

# Set display options to show full column content
pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

# Directory containing the extracted CSVs
data_dir = (
    r"D:\Data_Science\Projects\Portfolio_Projects\Retail-Merchandising-Pricing-Optimization-Customer-Engagement\Dataset\Raw\Extracted_Tables"
)
files = ["brands.csv", "finance.csv", "info.csv", "reviews.csv", "traffic.csv"]

for file in files:
    file_path = os.path.join(data_dir, file)
    df = pd.read_csv(file_path)

    print("=" * 60)
    print(f"INSPECTING FILE: {file}")
    print("=" * 60)

    # 1. Preview
    print("--- First 3 Rows Preview ---")
    print(df.head(3))
    print("\n" + "#" * 60 + "\n")

    # 2. Shape
    print(f"Dimensions: {df.shape[0]} rows, {df.shape[1]} columns\n")

    # 3. Schema and Data Types
    print("--- Data Types & Missing Values Summary ---")
    df.info()
    print("\n")

    # 4. Exact Missing Values Count
    print("--- Missing Values Count Per Column ---")
    print(df.isnull().sum())
    print("\n")

    # 5. Missing Values Percentage
    print("--- Missing Values Percentage ---")
    print((df.isnull().sum() / len(df)) * 100)
    print("\n")

    # 6. Duplicate Rows
    print(f"Total Duplicate Rows in Dataset: {df.duplicated().sum()}\n")

    # 7. Numerical Summary
    num_cols = df.select_dtypes(include=["number"])
    if not num_cols.empty:
        print("--- Numerical Summary ---")
        print(num_cols.describe())
        print("\n")

    # 8. Categorical Summary
    cat_cols = df.select_dtypes(include=["object", "category", "bool", "string"])
    if not cat_cols.empty:
        print("--- Categorical Summary ---")
        print(cat_cols.describe())
        print("\n")

    # 9. Uniqueness (Cardinality)
    print("--- Unique Values Count Per Column ---")
    print(df.nunique())
    print("\n")

    # 10. Value Counts for All Columns (Limited to Top 15 Results)
    print("--- Top 15 Most Frequent Value Counts Per Column ---")
    for col in df.columns:
        print(f"\nColumn: {col} (Total Unique: {df[col].nunique()})")
        print(df[col].value_counts().head(15))
    print("\n")
```

```python
import os
import sqlite3
import pandas as pd

# Define paths
db_path = (
    r"D:\Data_Science\Projects\Portfolio_Projects\Retail_Merchandising_Pricing_Optimization_And_Customer_Engagement\Dataset\Raw\retailDB.sqlite"
)
output_dir = (
    r"D:\Data_Science\Projects\Portfolio_Projects\Retail_Merchandising_Pricing_Optimization_And_Customer_Engagement\Dataset\Raw\Extracted_Tables"
)

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Connect to the SQLite database
conn = sqlite3.connect(db_path)

# Fetch all table names
tables = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table';", conn
)

# Export each table to a CSV file
for table_name in tables["name"]:
    df = pd.read_sql_query(f"SELECT * FROM [{table_name}]", conn)
    output_path = os.path.join(output_dir, f"{table_name}.csv")
    df.to_csv(output_path, index=False)
    print(f"Extracted: {table_name}.csv")

conn.close()
print("Extraction complete!")
```
